### Topic Covered: Data Cleaning Technique.

```python
#importing necessary packages
import pandas as pd
```

### Load the Toyota vehicle’s data from a given csv file into a dataframe and print it.  (Toyota.csv)


```python
df=pd.read_csv('Datasets/Toyota.csv')
df
```

### Replace all abnormal symbols such as ‘?’ ‘***’ etc. by Null values

```python
df.isnull()
```

```python
import numpy as np

df2= df.replace(to_replace =["??", "***","????"], 
                            value =np.NaN) 
df2
```

### Count the total number of null values present in each attribute (column) 


```python
print(df2.isnull().sum())
```

### Replace the NaN in Age column by median value.

```python
df2["Age"].fillna(df2["Age"].median(), inplace = True)
df2["Age"].median()
```

```python
df2
```

### Replace the string value in column Door with number.


```python
df2.replace({'Doors':{'three': 3,'four': 4, 'five' : 5}},inplace=True)
df2
```

```python
df2['Doors'].value_counts()
```

### Returns a Series containing counts of unique values for column ‘FuelType’ (Petrol, Diesel, CNG )


```python
df2['FuelType'].value_counts()
```

### Get the mode value of FuelType and  fill NA/NaN values using the specified value of mode.


```python
a = df2['FuelType'].mode()
a
```

```python
df2['FuelType'].fillna(df2['FuelType'].mode()[0], inplace=True)
display(df2)
```

### Get the mode value of Metcolor and  fill NA/NaN values using the specified value of mode.


```python
y = df2['MetColor'].mode()
y
```

```python
df2['MetColor'].fillna(df2['MetColor'].mode()[0], inplace=True)
display(df2)
```

### Replace the value of FuelType as for Petrol replaces with 0, diesel with 1 and CNG with 2.


```python
df2["FuelType"].replace({"Petrol": "0", "Diesel": "1","CNG":"2"}, inplace=True)
display(df2)
```

```python
df2['FuelType'].value_counts()
```

### Copy the dataset to  new dataframe and apply One hot encoding categorical approach on Fueltype column.

```python
df3=pd.DataFrame(df2)
df3
```

```python
df4 = pd.get_dummies(df3, columns = ['FuelType'])
print(df4)
```
