### View basic statistical details like percentile, mean, std deviation, mode, variance,skewness of the data in table & write this in dataset & further create CSV file.

```python
feild=['SGPA','CGPA','Marks(course1)','Marks(course1)','Marks(course1)']
mean=[]
std=[]
var=[]
skw=[]
mode=[]
for i in feild:
  mean.append(df[i].mean())
  std.append(df[i].std())
  var.append(df[i].var())
  skw.append(df[i].skew())
  mode.append(df[i].mode())
```

### Calculate the age of all the students from DOB & categorize it in the different age groups (such as between 15-20, 21-25 etc.)

```python
from datetime import date
#df['DOB'] = pd.to_datetime(df['DOB'],format='%d/%m/%Y')
#df['DOB']
```

```python
from datetime import datetime, date
def age(born):
    born = datetime.strptime(born, "%d/%m/%Y").date()
    today = date.today()
    return today.year - born.year - ((today.month, 
                                      today.day) < (born.month, 
                                                    born.day))
  
df['Age'] = df['DOB'].apply(age)
  
print(df['Age'])
```
