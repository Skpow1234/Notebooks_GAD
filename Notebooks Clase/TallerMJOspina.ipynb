```python
#IMPORTAR LIBRERIAS
import pandas as pd
import numpy as np
from sklearn.model_selection import train_test_split
from sklearn.datasets import load_iris
from sklearn import tree
from sklearn.ensemble import HistGradientBoostingClassifier
from sklearn.metrics import confusion_matrix
from sklearn.metrics import classification_report
```

```python
#PUNTO 1: Cargar conjunto de datos
df = pd.read_csv("games.csv")
df.head(10)
```

```python
#PUNTO 2: Particionar el conjunto
df.columns
```

```python
feactures = ['gameId', 'creationTime', 'gameDuration', 'seasonId','firstBlood', 'firstTower', 'firstInhibitor', 'firstBaron',
       'firstDragon', 'firstRiftHerald', 't1_champ1id', 't1_champ1_sum1','t1_champ1_sum2', 't1_champ2id', 't1_champ2_sum1', 't1_champ2_sum2',
       't1_champ3id', 't1_champ3_sum1', 't1_champ3_sum2', 't1_champ4id','t1_champ4_sum1', 't1_champ4_sum2', 't1_champ5id', 't1_champ5_sum1',
       't1_champ5_sum2', 't1_towerKills', 't1_inhibitorKills', 't1_baronKills','t1_dragonKills', 't1_riftHeraldKills', 't1_ban1', 't1_ban2',
       't1_ban3', 't1_ban4', 't1_ban5', 't2_champ1id', 't2_champ1_sum1', 't2_champ1_sum2', 't2_champ2id', 't2_champ2_sum1', 't2_champ2_sum2',
       't2_champ3id','t2_champ3_sum1', 't2_champ3_sum2', 't2_champ4id', 't2_champ4_sum1', 't2_champ4_sum2', 't2_champ5id', 't2_champ5_sum1',
       't2_champ5_sum2', 't2_towerKills', 't2_inhibitorKills', 't2_baronKills', 't2_dragonKills', 't2_riftHeraldKills', 't2_ban1', 't2_ban2',
       't2_ban3', 't2_ban4','t2_ban5']

X = df[feactures]
y = df['winner']
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)
```

```python
#PUNTO 3: Modelos de ML

#PRIMER MODELO: Decision Trees
clfDT = tree.DecisionTreeRegressor()
clfDT = clfDT.fit(X_train, y_train)
y_predDT = clfDT.predict(X_test)
```

```python
#SEGUNDO MODELO: Ensembles
clfE = HistGradientBoostingClassifier(max_iter=100)
clfE = clfE.fit(X_train, y_train)
y_predE = clfE.predict(X_test)
```

```python
#PUNTO 4: Metrica diagnositica classifation_report y confion_matrix

#PRIMER MODELO: Decision Trees
print(classification_report(y_test, y_predDT))
```

```python
confusion_matrix(y_test, y_predDT)
```

```python
#SEGUNDO MODELO: Ensembles
print(classification_report(y_test, y_predE))
```

```python
confusion_matrix(y_test, y_predE)
```
