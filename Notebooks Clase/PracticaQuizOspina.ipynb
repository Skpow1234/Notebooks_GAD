```python
#IMPORTAR LIBRERIAS
import pandas as pd #Procesación de datos
import numpy as np #Parte algebraica de los datos
import matplotlib.pyplot as plt #Parte grafica
```

```python
#CARGAR LA BASE DE DATOS
df = pd.read_csv("https://raw.githubusercontent.com/cgiohidalgo/repositoryclass/main/datatest/GAD_train_titanic.csv", sep = ";")
df.head(1308)
```

```python
df.isnull()
```

```python
#ELIMINA TODOS LAS COLUMNAS COMPLETAMENTE VACIAS
df = df.dropna(how='all')
df
```

```python
#LLENA LOS CAMPOS VACIOS DE LAS FILAS CON LOS DATOS DE LA FILA SIGUIENTE
# df = df.fillna(method='bfill')

```

```python
# df.fillna(value=0) # Cambia todos los nulos
df["Edad"].fillna(value=0, inplace=True)
df["Cabina"].fillna(value="T856", inplace=True)
df
```

```python
#SE ELIMINAN DOS COLUNMAS QUE NO SERAN MUY RELEVANTES
df = df.drop(['Hermanos-Esposos', 'Padres-Hijos'], axis=1)
df
```

```python
#GRAFICA DE LA CANTIDAD DE PERSONAS QUE SOBREVIVIERON SEGUN LA CLASE EN LA QUE VIAJARON

total = df.Sobrevive.sum()
plt.bar(df.Clase, df.Sobrevive)
plt.show()
```
